# IMPORTACIÓN DATOS
Importamos la librería pandsas para guardar y tratar el fichero train.csv

```python
import pandas as pd
train_dataframe= pd.read_csv("https://raw.githubusercontent.com/rodrigotorresp/Practica-Regresi-n-ML/main/train.csv");
train_dataframe
```

# PREPROCESAMIENTO

Uno de los problemas que existen con el tratamiento de estos datos es la existencia de las unidades de medida en algunas columnas, es decir, cuando hablamos de peso el string "kg" está presente en todos los valores del dataset, de manera que es imprescindible eliminarlo de primeras.

De la misma forma, con la columna CPU intentamos separar las unidades de medida utilizando el replace, sustituyendo los espacios en blanco por guiones. Nuestro siguiente paso sería dividir con el método split, en columnas lo que ha sido separado por guiones.

Nuestra intención era utilizar la medida de la frecuencia en GHz, pero al quedarse en algunos casos en columnas distintas no nos fue posible. Sin embargo, pudimos utilizar la marca de procesadores y hacer con ello posteriormente un ordinal encoder, pero no mejoró la predicción.

```python
train_dataframe['Weight'] = train_dataframe['Weight'].str.replace('kg$', '')

train_dataframe['CPU2'] = train_dataframe['Cpu'].str.replace('\s', '-')
cpu = train_dataframe['CPU2'].str.split('[-]',expand=True)

train_dataframe=pd.concat([train_dataframe, cpu], axis=1)
```

```python
train_dataframe
```

  - Eliminamos LaptopId ya que no nos aporta ningun valor el tener el identificador del portátil.

  - Podríamos haber tenido en cuenta la compañía que fabrica el dispositvo pero revisando los datos, hay gran varianza en los precios de las diferentes compañías.

  - La columna 'Product' decidimos prescindir de ella, ya que considerabamos que no aportaba información.

  - Con la columna 'Typename' realimos un ordinalencoder, dándole más importancia a 'Workstation' que a las demás. Posteriormente aplicaremos también un StandardScaler, para ajustar los datos y que el modelo pueda aprender bien.

  - Con respecto a 'Inches', en la primera transformación pasamos de largo, pero en la segunda aplicamos también un StandardScaler.

  - 'ScreenResolution' se intentó proceder de la misma manera que con Cpu (antes comentado) y al no conseguir el valor numérico de la resolución en una misma columna fue imposible utilizar está información.

  - Eliminamos la columna Cpu, utilizaremos las columnas que concatenamos en el anterior bloque de código.

  - Con la memoria RAM utilizamos un OrdinalEncoder por categorías, yendo de mayor a menor prioridad desde 2GB hasta 64GB.

  - Tuvimos la idea de utilizar la información de la columna 'Memory' de manera que tuviesemos una columna que nos indicase con un 1 o un 0 si la memoria era de tipo HDD o de tipo SSD, utilizando la sentencia OneHotEncoder, pero no fuimos capaces de aislar en una misma columna el string 'HDD' o 'SSD' para utilizarlo. Al final la eliminamos

  - Con la columna 'GPU' nos encontramos con el mismo problema.

  - Utilizamos de nuevo el OrdinalEncoder con categorías para ordenar por prioridad el sistema operativo. Posteriormente aplicaremos StandardScaler.

  - En el caso de 'Weight' no realiazamos ningun cambio en el primer bloque de código, para posteriormente usar StandardScaler.

  - Eliminamos 'Price' porque es nuestra variable objetivo, por tanto no debe existir en el conjunto de entrenamiento.

  - 'CPU2' se elimina porque es el vector separado por guiones que utilizamos para separar por columnas.

  - 'INTEL/AMD' sería la columna que usaríamos para hacer un ordinalEnconder con Intel AMD y Samsung, pero decidimos no usarlo ya que no mejoraba las predicciones.

  - Las demás columnas se eliminaron puesto que no pudimos asilar en una misma columna los datos necesarios para usar esa información (antes explicado)

```python
import sklearn.compose
colum_transformer = sklearn.compose.ColumnTransformer(transformers=[
    ("Laptop Id" , "drop", [0]),
    ("Company" , "drop", [1]),
    ("Product" , "drop", [2]),
    ("TypeName" , sklearn.preprocessing.OrdinalEncoder(categories=[['Netbook', 'Notebook', 'Ultrabook','Gaming','2 in 1 Convertible','Workstation']]), [3]),
    ("Inches" , "passthrough", [4]),
    ("ScreenResolution" , "drop", [5]),
    ("CPU" , "drop", [6]),
    ("RAM" , sklearn.preprocessing.OrdinalEncoder(categories=[['2GB', '4GB','6GB','8GB','12GB','16GB','24GB','32GB' ,'64GB']]), [7]),
    ("Memory" , "drop", [8]),
    ("GPU" , "drop", [9]),
    ("OPSYS" , sklearn.preprocessing.OrdinalEncoder(categories=[['No OS', 'Android' , 'Chrome OS' , 'Windows 7','Linux','Windows 10','macOS','Windows 10 S','Mac OS X']]), [10]),
    ("Weight" , "passthrough", [11]),
    ("Price" , "drop", [12]),
    ("CPU2", "drop", [13]),
    ("INTEL/AMD", "drop", [14]),
    ("1", "drop", [15]),
    ("2", "drop", [16]),
    ("3", "drop", [17]),
    ("4", "drop", [18]),
    ("5", "drop", [19]),
    ]);

train_dataframe_transform = colum_transformer.fit_transform(train_dataframe)
train_dataframe_transform
```

```python
import sklearn.compose
colum_transformer_Scaler = sklearn.compose.ColumnTransformer(transformers=[
    ("TypeNameScaler" , sklearn.preprocessing.StandardScaler(), [0]),
    ("InchesScaler" , sklearn.preprocessing.StandardScaler(), [1]),
    ("RAMScaler" , sklearn.preprocessing.StandardScaler(), [2]),
    ("WeightScaler" , sklearn.preprocessing.StandardScaler(), [4]),
    ("OPSYSScaler" , sklearn.preprocessing.StandardScaler(), [3]),
    ]);

train_dataframe_transform = colum_transformer_Scaler.fit_transform(train_dataframe_transform)
train_dataframe_transform
```

En primera instancia, pensamos en una regresión lineal, ya que no teníamos muchos parámetros de entrada y parecía viable a primera vista. 

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
 
X = train_dataframe_transform.data
y = train_dataframe.Price

X_train = train_dataframe_transform.data
y_train = train_dataframe.Price

reg = LinearRegression().fit(X, y)
reg.intercept_
```

Su valor es 1130.0370065789473 que representa el precio básico de cualquier portátil sin contar con el resto de las características que describan los portátiles.

Ya habiendo entrenado la regresión lineal 'reg' revisamos sus métricas de evaluacíon

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

y_pred = reg.predict(X_train)
print("MAE: ", mean_absolute_error(y_train, y_pred))
print("MSE: ", mean_squared_error(y_train, y_pred))
print("R2: ", reg.score(X_train, y_train))
```

Hicimos una primera submission y decidimos probar otros modelos para comparar y ver si alguno lo mejoraba

Usamos Lasso porque sospechabamos que varios de los atributos de entrada serían irrelevantes, pero no generó buenos resultados.

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0).fit(X_train,y_train)
y_pred = lasso.predict(X_train)
print("MAE: ", mean_absolute_error(y_train, y_pred))
print("MSE: ", mean_squared_error(y_train, y_pred))
print("R2: ", lasso.score(X_train, y_train))

lasso2 = Lasso(alpha=2.5).fit(X_train,y_train)
y_pred = lasso.predict(X_train)
print("MAE: ", mean_absolute_error(y_train, y_pred))
print("MSE: ", mean_squared_error(y_train, y_pred))
print("R2: ", lasso2.score(X_train, y_train))

lasso5 = Lasso(alpha=5.0).fit(X_train,y_train)
y_pred = lasso.predict(X_train)
print("MAE: ", mean_absolute_error(y_train, y_pred))
print("MSE: ", mean_squared_error(y_train, y_pred))
print("R2: ", lasso5.score(X_train, y_train))

lasso7 = Lasso(alpha=7.5).fit(X_train,y_train)
y_pred = lasso.predict(X_train)
print("MAE: ", mean_absolute_error(y_train, y_pred))
print("MSE: ", mean_squared_error(y_train, y_pred))
print("R2: ", lasso7.score(X_train, y_train))
```

Dado que Lasso dió malos resultados,optamos por probar con una regresión lineal pero con regularización, como lo es Ridge. Ya que requiere que los atributos este correlados entre si.Aplicamos distintos valores al hiper-parámetro alpha para ver cual producía el menor error.

```python
from sklearn.linear_model import Ridge

ridge1 = Ridge(alpha=1.0).fit(X_train, y_train)
y_pred = ridge1.predict(X_train)
print("MAE: ", mean_absolute_error(y_train, y_pred))
print("MSE: ", mean_squared_error(y_train, y_pred))
print("R2: ", ridge1.score(X_train, y_train))

ridge5 = Ridge(alpha=5.0).fit(X_train, y_train)
y_pred = ridge5.predict(X_train)
print("MAE: ", mean_absolute_error(y_train, y_pred))
print("MSE: ", mean_squared_error(y_train, y_pred))
print("R2: ", ridge5.score(X_train, y_train))

ridge10 = Ridge(alpha=10.0).fit(X_train, y_train)
y_pred = ridge10.predict(X_train)
print("MAE: ", mean_absolute_error(y_train, y_pred))
print("MSE: ", mean_squared_error(y_train, y_pred))
print("R2: ", ridge10.score(X_train, y_train))

ridge15 = Ridge(alpha=15.0).fit(X_train, y_train)
y_pred = ridge15.predict(X_train)
print("MAE: ", mean_absolute_error(y_train, y_pred))
print("MSE: ", mean_squared_error(y_train, y_pred))
print("R2: ", ridge15.score(X_train, y_train))
```

Tembién probamos la regresión de Bayes pero no generó mejores resultados

```python
from sklearn import linear_model

Bayes = linear_model.BayesianRidge()
Bayes.fit(X_train, y_train)
y_pred = Bayes.predict(X_train)

print("MAE: ", mean_absolute_error(y_train, y_pred))
print("MSE: ", mean_squared_error(y_train, y_pred))
print("R2: ", Bayes.score(X_train, y_train))
```

Usamos la regresión tree, que es buen modelo a la hora de predecir una variable a partir de variables independientes entre si, pero resulto generar overfitting

```python
from sklearn import tree
tree = tree.DecisionTreeRegressor()
tree = tree.fit(X_train, y_train)
y_pred = tree.predict(X_train)

print("MAE: ", mean_absolute_error(y_train, y_pred))
print("MSE: ", mean_squared_error(y_train, y_pred))
print("R2: ", tree.score(X_train, y_train))
```

Al usar la regresión KNeighbors dió mejores resultados
Los parametros significan:
- n_neighbors: Número de vecinos que se usarán de forma predeterminada para las consultas de kneighbors.
- weights: Ponderaciones. Al usar 'uniform' todos los puntos en cada vecindario se ponderan por igual.

```python
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(n_neighbors=10, weights='uniform').fit(X_train, y_train)
y_pred = knn.predict(X_train)
print("MAE: ", mean_absolute_error(y_train, y_pred))
print("MSE: ", mean_squared_error(y_train, y_pred))
print("R2: ", knn.score(X_train, y_train))
```
